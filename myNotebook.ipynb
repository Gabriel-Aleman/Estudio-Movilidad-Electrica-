```python
import pandas as pd
import numpy as np

def montecarlo_from_df(df, date_col="t_stamp", value_col=None, dias_futuros=365, n_simulaciones=1000):
    """
    Aplica Montecarlo a una columna numÃ©rica basada en series de tiempo.
    
    ParÃ¡metros:
    -----------
    df : DataFrame con una columna de fechas y varias columnas de valores.
    date_col : str -> columna tipo datetime.
    value_col : str -> columna a analizar (consumo).
    dias_futuros : int -> dÃ­as a simular.
    n_simulaciones : int -> cantidad de corridas del modelo.
    
    Retorna:
    --------
    DataFrame con fecha futura, promedio, p5, p95.
    """

    if value_col is None:
        raise ValueError("Debes especificar la columna de consumo usando `value_col`.")

    df = df.copy()

    # ---- 1. Convertir fecha y ordenar ----
    df[date_col] = pd.to_datetime(df[date_col])
    df = df.sort_values(date_col)

    # ---- 2. Calcular retornos porcentuales ----
    df['retorno'] = df[value_col].pct_change()
    retornos = df['retorno'].dropna().values

    if len(retornos) == 0:
        raise ValueError("No se pueden calcular retornos. Necesitas al menos 2 registros vÃ¡lidos.")

    # ---- 3. Ãšltimo valor real ----
    valor_actual = df[value_col].iloc[-1]

    # ---- 4. Matriz simulaciones ----
    simulaciones = np.zeros((dias_futuros, n_simulaciones))

    for s in range(n_simulaciones):
        valor = valor_actual
        for d in range(dias_futuros):
            r = np.random.choice(retornos)
            valor = valor * (1 + r)
            simulaciones[d, s] = valor

    # ---- 5. Crear fechas futuras ----
    fechas_futuras = pd.date_range(start=df[date_col].iloc[-1] + pd.Timedelta(days=1), periods=dias_futuros)

    # ---- 6. Resultado final ----
    pred_df = pd.DataFrame({
        'fecha': fechas_futuras,
        'promedio': simulaciones.mean(axis=1),
        'p5': np.percentile(simulaciones, 5, axis=1),
        'p95': np.percentile(simulaciones, 95, axis=1),
    })

    return pred_df
montecarlo_from_df(df, value_col="Consumo total (kW)")
```

```python
import pandas as pd
import numpy as np


def montecarlo_predict(df, n_simulaciones=1000, dias_futuros=365, date_col="t_stamp", value_col="Consumo total (kW)"):
    """
    Recibe un DataFrame con columnas:
        - date_col -> formato como 'Mon 00:00:00' (sin t_stamp real completa)
        - value_col -> valores numÃ©ricos (consumo/mediciÃ³n)

    Reconstruye t_stamps consecutivas automÃ¡ticamente, calcula retornos histÃ³ricos
    y genera predicciÃ³n Montecarlo para `dias_futuros`.

    Retorna un DataFrame con:
        t_stamp, promedio, p5 (peor caso), p95 (mejor caso).
    """

    df = df.copy()

    # ---- 1. Reconstruir t_stamps reales porque vienen incompletas ----
    df[date_col] = df[date_col].astype(str)
    df = df.sort_index()

    # Creamos t_stamps desde el dÃ­a actual hacia adelante segÃºn tamaÃ±o del dataset
    df["t_stamp"] = pd.date_range(start=pd.Timestamp.today().normalize(), periods=len(df))

    # ---- 2. Calcular retornos histÃ³ricos (variaciones porcentuales) ----
    df['retorno'] = df[value_col].pct_change()
    retornos = df['retorno'].dropna().values

    if len(retornos) == 0:
        raise ValueError("No hay suficientes datos para calcular retornos.")

    # ---- 3. Ãšltimo valor real para iniciar simulaciÃ³n ----
    valor_actual = df[value_col].iloc[-1]

    # ---- 4. Matriz de simulaciones ----
    simulaciones = np.zeros((dias_futuros, n_simulaciones))

    for s in range(n_simulaciones):
        valor = valor_actual
        for d in range(dias_futuros):
            r = np.random.choice(retornos)  # muestra un cambio real histÃ³rico
            valor = valor * (1 + r)
            simulaciones[d, s] = valor

    # ---- 5. Crear t_stamps futuras ----
    t_stamps_futuras = pd.date_range(start=df['t_stamp'].iloc[-1] + pd.Timedelta(days=1), periods=dias_futuros)

    # ---- 6. Ensamblar resultado ----
    pred_df = pd.DataFrame({
        't_stamp': t_stamps_futuras,
        'promedio': simulaciones.mean(axis=1),
        'p5': np.percentile(simulaciones, 5, axis=1),
        'p95': np.percentile(simulaciones, 95, axis=1)
    })

    return pred_df
montecarlo_predict(df)
```
